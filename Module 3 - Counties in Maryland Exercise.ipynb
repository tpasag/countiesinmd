```python
import requests #library used to connect to a website
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland#List_of_counties"
```

```python
page = requests.get(URL)
```

```python
page.status_code
```

```python
page.text
```

```python
HTMLstr = page.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(HTMLstr, "lxml")
```

```python
print(soup.title.string)
```

```python
use_table=soup.find('table', class_='wikitable sortable')
use_table
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]

for row in use_table.findAll("tr"):
    cells1 = row.findAll('th')
    if len(cells1)==1:
        A.append(cells1[0].find(text=True))
        
    cells = row.findAll('td')
    if len(cells)==10:
        B.append(cells[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[7].find(text=True))
        H.append(cells[8].find(text=True))
```

```python
A
```

```python
H
```

```python
#import pandas to convert list to data frame
import pandas as pd

df=pd.DataFrame(A, columns=['County']) #turn list A into dataframe first

#add other lists as new columns
df['FIPS Code'] = B
df['CountySeat'] = C
df['Established'] = D
df['Origin'] = E
df['Entymology'] = F
df['Population'] = G
df['Area'] = H

#show first 5 rows of created dataframe
df.head()
```

```python
#export scraped data to a csv file
df.to_csv("CapitalList.csv")
```
